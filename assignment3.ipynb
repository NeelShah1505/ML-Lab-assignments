```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#import data (heart.csv)
df = pd.read_csv('/content/drive/MyDrive/ml assigments/datasets/Heart.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='MaxHR', hue='ChestPain', data=df, palette=['pink', 'yellow', 'lightblue', 'purple'])
plt.title('Scatter Plot')
plt.xlabel('Age')
plt.ylabel('MaxHR')
plt.show()
```

```python
# implementing bar plot
plt.figure(figsize=(8, 6))
df['ChestPain'].value_counts().plot(kind='bar', color='blue')
plt.title('Bar Plot: ChestPain Frequency')
plt.xlabel('ChestPain Type')
plt.ylabel('Frequency')
plt.show()
```

```python
# implementing Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(y='Age', data=df, color='blue')
plt.title('Box Plot')
plt.ylabel('Age')
plt.show()
```

```python


 # implementing pie chart
plt.figure(figsize=(8, 6))
colors = ['red', 'yellow', 'lightblue', 'gold']
plt.pie(df['Thal'].value_counts(), labels=df['Thal'].value_counts().index, autopct='%1.1f%%', colors=colors)
plt.title('Pie Chart')
plt.ylabel('')
plt.show()
```

```python
# implementing line chart
AgeArr = np.array(df['Age'])
SortedAgeArr = np.sort(AgeArr)
plt.figure(figsize=(24, 6))
plt.plot(SortedAgeArr, df['Chol'], marker='o', linestyle='-', color='red')  # Setting line color to pink
plt.title('Line Chart: Age vs. Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol (Chol)')
plt.grid(True)
plt.show()
```
